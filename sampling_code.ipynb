```python
import os
import json
import random

# Path to the zero-shot directory
# folder_path = "zero-shot"
folder_path = "few-shot"
sample_size = 20

# Loop over each JSON file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".json") and not filename.endswith("_sample.json"):
        file_path = os.path.join(folder_path, filename)

        # Read the JSON data
        with open(file_path, "r") as f:
            data = json.load(f)

        # Sample 20 elements (or fewer if not enough data)
        sample = random.sample(data, min(sample_size, len(data)))

        # Save to a new file
        sample_file_path = os.path.join(folder_path, filename.replace(".json", "_sample.json"))
        with open(sample_file_path, "w") as f:
            json.dump(sample, f, indent=2)

        print(f"Saved sample to {sample_file_path}")
```
